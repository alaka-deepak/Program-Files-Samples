### Decision Tree
Supervised learning technique that can be used for both <b>classification and Regression </b>problems.
<li>Preferred for solving Classification problems. 
<li>It is a <b>Tree-structured classifier,</b>
<li>where <b>Internal nodes</b> represent <b>the features of a dataset,</b> <li><b>Branches </b>represent the <b>Decision rules</b> and 
<li><b>Leaf node </b>represents <b>the outcome.</b>

  
Two nodes,
<li><b>Decision Node
    <li>Leaf Node</b>
<li>It is a graphical representation for getting all the possible solutions to a problem/decision based on given conditions.       
<li>It is called a decision tree because, similar to a tree, it starts with the root node, which expands on further branches and constructs a tree-like structure.
    In order to build a tree, we use the CART algorithm, which stands for Classification and Regression Tree algorithm.

<b> A decision tree simply asks a question, and based on the answer (Yes/No), it further split the tree into subtrees.</b> 
    

### Decision Tree Terminologies
<b><li>Root Node: </b>Root node is from where the decision tree starts. It represents the entire dataset, which further gets divided into two or more homogeneous sets.
<b><li>Leaf Node:</b> Leaf nodes are the final output node, and the tree cannot be segregated further after getting a leaf node.
<b><li>Splitting:</b> Splitting is the process of dividing the decision node/root node into sub-nodes according to the given conditions.
<b><li>Branch/Sub Tree: </b>A tree formed by splitting the tree.
<b><li>Pruning:</b>Pruning is the process of removing the unwanted branches from the tree.
<b><li>Parent/Child node:</b> The root node of the tree is called the parent node, and other nodes are called the child nodes.

### Algorithm


    Step-1: Begin the tree with the root node, says S, which contains the complete dataset.
    Step-2: Find the best attribute in the dataset using Attribute Selection Measure (ASM).
    Step-3: Divide the S into subsets that contains possible values for the best attributes.
    Step-4: Generate the decision tree node, which contains the best attribute.
    Step-5: Recursively make new decision trees using the subsets of the dataset created in step -3. 
    Continue this process until a stage is reached where you cannot further classify the nodes and called the
    final node as a leaf node.



<b>Attribute Selection Measures</b>
    
   Can easily select the best attribute for the nodes of the tree. 
    

There are two popular techniques for ASM, which are:    
   <li> Information Gain
   <li> Gini Index


<b>1.Information gain</b>
<li><b>Measurement of changes in entropy</b> after the segmentation of a dataset based on an attribute.
<li>It calculates how much information a feature provides us about a class. 
<li>According to the value of information gain,<b>split the node and build the decision tree. </b>
<li>A node/attribute having <b>the highest information gain is split first.</b>
<li>It can be calculated using the below formula:

<b> Information Gain= Entropy(S)- [(Weighted Avg) *Entropy(each feature)] </b>

<b>2.Entropy: </b>Entropy is a metric to measure the impurity in a given attribute. It specifies randomness in data.

<b>Entropy(s)= -P(yes)log2 P(yes)- P(no) log2 P(no)</b>
Where,

    S= Total number of samples
    P(yes)= probability of yes
    P(no)= probability of no
    
    
<b>3. Gini Index:</b>

<li><b>Gini index is a measure of impurity or purity </b>used while creating a decision tree in the <b>CART(Classification and Regression Tree)</b> algorithm. 

<li>The low Gini index should be preferred for an attribute.

<li>It only creates binary splits.

<b>Gini Index= 1- (P)2</b>

<b>4.Pruning:</b> Getting an Optimal Decision tree

<li>Pruning is a process of deleting the unnecessary nodes from a tree in order to get the optimal decision tree. 

<li>A too-large tree increases the risk of overfitting <li>Small tree may not capture all the important features of the dataset. 

<li>Therefore, a technique that decreases the size of the learning tree without reducing accuracy is known as <b>Pruning. </b>
    <li>There are mainly two types of tree pruning technology used:

<b>
    Cost Complexity Pruning,
    Reduced Error Pruning.</b>



### Advantages 

   <li> Simple to understand -follows the same process which a human follow while making any decision in real-life.
<li>It can be very useful for solving decision-related problems.
<li>It helps to think about all the possible outcomes for a problem.
<li>There is less requirement of data cleaning compared to other algorithms.
    
### Disadvantages 

<li>The decision tree contains lots of layers, which makes it complex.
<li>It may have an overfitting issue, which can be resolved using <b>the Random Forest algorithm.</b>
<li>For more class labels, the computational complexity of the decision tree may increase.


### Random Forest Classifier 

<li>Process of combining multiple classifiers to solve a complex problem and to improve the performance of the model.
<li>Random Forest is a classifier that contains a number of decision trees on various subsets of the given dataset and takes the average to <b>improve the predictive accuracy of that dataset.</b>
<li>Based on the concept of <b>ensemble learning.</b>

<li>Random forest takes the prediction from each tree and based on the majority votes of predictions, and it predicts the final output. 

<li>The greater number of trees in the forest leads to higher accuracy and prevents the problem of overfitting. 




<b>The predictions from each tree must have very low correlations.</b>
    
### Why use Random Forest?    

    
<li>It takes less training time as compared to other algorithms.
<li>It predicts output with high accuracy, even for the large dataset it runs efficiently.
<li>It can also maintain accuracy when a large proportion of data is missing. 

    
<b>Advantages of Random Forest</b>

 <li>   Random Forest is capable of performing both Classification and Regression tasks.
  <li>  It is capable of handling large datasets with high dimensionality.
 <li>   It enhances the accuracy of the model and prevents the overfitting issue.


    

### Random Forests vs Decision Trees

   <li> Random forests is a set of multiple decision trees.
  <li>  Deep decision trees may suffer from overfitting, but random forests prevents overfitting by creating trees on random subsets.
  <li>  Decision trees are computationally faster.
   <li> Random forests is difficult to interpret, while a decision tree is easily interpretable and can be converted to rules.
