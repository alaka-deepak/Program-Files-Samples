## Iterators

- An iterator is an object that contains a countable number of values.

- An iterator is an object that can be iterated upon, meaning that you can traverse through all the values.

- Technically, in Python, an iterator is an object which implements the iterator protocol, which consist of the methods __iter__() and __next__().

- Lists, tuples, dictionaries, and sets are all iterable objects. They are iterable containers which you can get an iterator from.
- allow for efficient looping and memory management

```python
myitem=("abc","def","ghi")
myit=iter(myitem)

print(next(myit))
print(next(myit))
```

```python
# Strings are also iterable objects, containing a sequence of characters:

myst="banana"
mysit=iter(myst)
print(next(mysit))
print(next(mysit))
print(next(mysit))
```

```python
# Looping through an iterator

mytuple=("apple","banana","mango")
for x in mytuple:
    print(x)
```

```python
# Create an Iterator

# __init__ method acts similar , some initializing when the object is being created.

class Mynumbers:
    def __iter__(self):
        self.a=1
        return self
    
    def __next__(self):
        x=self.a
        self.a +=1
        return x
```

### Generators

- Generators are simpler way to create iterators. 
- Yield keyword generate value on fly and do not store in the memory
- A generator function is a special type of function that returns an iterator object. 
- Instead of using return to send back a single value,
- Generator functions use yield to produce a series of results over time. 
- The function pauses its execution after yield, maintaining its state between iterations.

```python
def square(n):
    for i in range(3):
        yield i**2

print(square(2))
```

```python
for i in square(3):
    print(i)
```

### Need of Generators

1. Memory Efficient
2. No List Overhead
3. Supported infinte sequences. (generating unbounded data)
4. Pipeline processing (pprocess data in stages efficiently)
5. Lazy Evaluation.
6. Generators are especially useful for processing large data files, like logs, because they handle data in small parts, saving memory and we do not have to load the entire file at once
7. 

```python
def my_generators():
    yield 1
    yield 2
    yield 3
```

```python
gen=my_generators()
gen
```

```python
for i in gen:
    print(i)
```

```python
# Reading large files

# without loading entire file into the memory, allow process one line at a time

def read_file(file_path):
    with open(file_path,'r') as file:
        for line in file:
            yield line

file_path='destination.txt'
```

```python
# Read line by line
for i in read_file(file_path):
    print(i.strip())
```

### Decorators

1. Modify or extend behavior of functions or methods, without changing their actual code
2. Decorators are often used in scenarios such as logging, authentication and memorization, allowing us to add additional functionality to existing functions or methods in a clean, reusable way.


```python

def welcome():
    return "Welcome to blast the Python"

welcome()
```

```python
wel=welcome
print(wel())
del welcome
print(wel())
```

```python
### closure functions 

def main_welcome(leng,lst):
    msg="welcome"
    def sub_welcome():
        print("Advanced python concepts")
        print(leng(lst))
        print(msg)
        print("Please do well")
    return sub_welcome()
```

```python
main_welcome(len,[1,2,3,4,5,6])
```

```python
def function_deco(func):
    def wrapper():
        print("Before calling the function")
        func()
        print("After calling the function")
    return wrapper

@function_deco # apply decorator to a function
def greet():
    print("Hello World")

greet()

# decorator creaate greet as an argument
# @decorator=decorator(greet)
```

### Decorator with Parameters

1. Decorators often need to work with functions that have arguments. 
2. We use *args and **kwargs so our wrapper can accept any number of arguments.

```python
def decorator_name(func):
    def wrapper(self):
        print("decorator checking")
        res=func(self)
        print("Second checking")
        return res
    return wrapper
```

```python
class Method:
    @decorator_name
    def say_hello(self):
    
        print("Hello")

    # say_hello=decorator_name(say_hello)

obj=Method()
obj.say_hello()
```

```python
def args_sample(fun):
    def wrapper(self,*args,**kwargs):
        print("first",args,kwargs)
        fun(self,*args,**kwargs)
        print("second")
    return wrapper

```

```python
class Method:
    @args_sample
    def say_hello(self,a,b,d,e,f,c):
    
        print("Hello",a,b,d,e,f,c)

    # say_hello=decorator_name(say_hello)

obj=Method()
obj.say_hello(10,20,45,43,55,c=30)
```
