### 1.Barplot
Barplots are meant for plotting categorical columns vs numerical columns. It creates bar in visualisation. Lets create a barplot of “total_bill” with “sex” and let’s see who pays more Males or Females.


```python
sns.barplot(x ='sex', y ='total_bill',data = tips)
```

```python
sns.barplot(x ='sex', y ='total_bill', hue ='smoker',data = tips)
```

```python
sns.barplot(x ='sex', y ='total_bill', data = tips, palette ='bone')
```

```python
sns.set_style('whitegrid')
sns.barplot(x ='sex', y ='total_bill', data = tips, palette ='rainbow')
```

#Inference - Total Bill Amount for males is more than Females.

### 2. Countplot
A countplot basically counts the categories and returns a count of their occurrences. It is one of the most simple plots provided by the seaborn library.

```python
sns.countplot(x ='sex', data = tips)
```

### 3.Boxplot
A Box Plot is a visual representation of five point summary statistics of a given data set. A five number summary includes:
Minimum,
First Quartile,
Median (Second Quartile),
Third Quartile,
Maximum
boxplot is created for Categorical — Continuous Variables which means that if the x -axis is categorical and y axis is continuous then a boxplot or a violin plot should be created.


```python
sns.boxplot(x ='day', y ='total_bill', data = tips)
```

```python
sns.boxplot(x ='day', y ='total_bill', data = tips, hue ='smoker')
```

#On Friday people have more bill if they are a Non smoker vs smoker

### 4.Violinplot
It is similar to the boxplot except that it provides a higher, more advanced visualization and uses the kernel density estimation to give a better description about the data distribution

```python
sns.violinplot(x ='day', y ='total_bill', data = tips, hue ='sex', split = False)
```

### 5.Stripplot
It basically creates a scatter plot based on the category

```python
sns.stripplot(x ='day', y ='total_bill', data = tips,
              jitter = True, hue ='smoker', dodge = False)
```

### 6.Swarmplot
It is very similar to the stripplot except the fact that the points are adjusted so that they do not overlap.Some people also like combining the idea of a violin plot and a stripplot to form this plot. One drawback to using swarmplot is that sometimes they dont scale well to really large numbers and takes a lot of computation to arrange them. So in case we want to visualize a swarmplot properly we can plot it on top of a violinplot.

```python
sns.swarmplot(x ='day', y ='total_bill', data = tips)
```

## Distribution Plots
The distribution plots in seaborn which is used for examining univariate and bivariate distributions.These plots are:

<ol>
    <li> distplot
<li>jointplot
<li>pairplot
<li>rugplot
<li>kdeplot

### 1.Distplot

It is used basically for univariate set of observations and visualizes it through a histogram i.e. only one observation and hence we choose one particular column of the dataset.sns.distplot() combines the histogram & plots the estimated probability density function over the data.

```python
sns.distplot(tips['total_bill'])
```

```python
# set the background style of the plot
sns.set_style('darkgrid')
sns.distplot(tips['total_bill'], kde = False, color ='red', bins = 30)
```

### 2.Jointplot
A Joint-Plot takes two variables and creates Histogram and Scatterplot together.

```python
sns.jointplot(x ='total_bill', y ='tip', data = tips, kind ='kde')
```

```python
sns.jointplot(x ='total_bill', y ='tip', data = tips, kind ='kde',shade= 'True' ,cmap= 'rainbow')
```

```python
sns.jointplot(x ='total_bill', y ='tip', data = tips, kind ='hex')
```

### 3.Pairplot
A Pairplot essentially plots pair wise relationship between variables. It supports “hue” as coloring the plot using categorical variables.


```python
sns.pairplot(tips, hue ="sex", palette ='rainbow')
```

### 4.Rugplot
It plots datapoints in an array as sticks on an axis.Just like a distplot it takes a single column. Instead of drawing a histogram it creates dashes all across the plot. If you compare it with the joinplot you can see that what a jointplot does is that it counts the dashes and shows it as bins.

```python
sns.rugplot(tips['total_bill'])
```

### 5.kdeplot
KDE Plot described as Kernel Density Estimate is used for visualizing the Probability Density of a continuous variable. It depicts the probability density at different values in a continuous variable.A KDE plot smooths the observations with a Gaussian kernel, producing a continuous density estimate. We can also plot a single graph for multiple samples which helps in more efficient data visualization.

```python
sns.kdeplot(data=tips, x="total_bill")
```

```python
sns.kdeplot(data=tips, y="total_bill")
```

```python
sns.kdeplot(data=tips, x="total_bill", hue="time")
```

```python
sns.kdeplot(tips['total_bill'])
sns.rugplot(tips['total_bill'])
```

## Relational plots in seaborn

Relational plots are used for visualizing the statistical relationship between the data points,ie ,how the variables in the dataset relate each other and their relationships

<ol>

<li>rel plot
<li>scatter plot
<li>line plot


### 1. relplot()

Seaborn.relplot() provides the access to some other different axes-level functions which shows the relationships between two variables with semantic mappings of subsets.

```python
# plotting a simple visualization of data points

sns.relplot(x ="total_bill", y ="tip", data = tips)
```

#### Example1: Grouping data points on the basis of category, here as time

```python
sns.relplot(x="total_bill",
            y="tip",
            hue="time",
            data=tips)
```

####  Example 2: Using relplot() with kind=”line”

```python

sns.relplot(x ="total_bill",
            y ="tip",
            kind ="line",
            data = tips)
```

#### Example 3: using time and sex for determining the facet of the grid

```python
sns.relplot(x="total_bill",
            y="tip",
            hue="day",
            col="time",
            row="sex",
            data=tips)
```

#### Example 4: using size attribute, we can see data points having different size

```python
sns.relplot(x="total_bill",
            y="tip",
            hue="day",
            size="size",
            data=tips)
```

### 2. scatterplot()
The scatter plot depicts the joint distribution of two variables using a cloud of points, where each point represents an observation in the dataset. This depiction allows the eye to infer a substantial amount of information about whether there is any meaningful relationship between them.



```python
# plotting a simple visualization of data points

sns.scatterplot(x ="total_bill", y ="tip", data = tips)
```

#### Example 1: Plotting a scatterplot using marker to differentiate between timing of the people visiting the restaurant.

```python
markers = {"Lunch": "s", "Dinner": "X"}

ax = sns.scatterplot(x ="total_bill",
                     y ="tip",
                     style ="time",
                     markers = markers,
                     data = tips)
```

### 3. lineplot()
Scatter plots are highly effective, but there is no universally optimal type of visualization. For certain datasets, you may want to consider changes as a function of time in one variable, or as a similarly continuous variable. In this case, drawing a line-plot is a better option.

```python
sns.lineplot(x ="total_bill", y ="tip", data = tips)
```

```python
sns.set(style = 'whitegrid')
sns.lineplot(x ="total_bill", y ="tip", data = tips)
```
