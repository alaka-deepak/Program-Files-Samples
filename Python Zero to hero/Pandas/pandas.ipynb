### Pandas Series
A Pandas Series is like a column in a table.

It is a one-dimensional array holding data of any type.

Create a simple Pandas Series from a list:


```python
a = [1, 7, 2]
myvar = pd.Series(a)
```

1. If nothing else is specified, the values are labeled with their index number. First value has index 0, second value has index 1 etc.

2. This label can be used to access a specified value.with the index argument, you can name your own labels.when you have created labels, you can access an item by referring to the label.


```python

myvar = pd.Series(a, index = ["x", "y", "z"])
print(myvar)
```

#### You can also use a key/value object, like a dictionary, when creating a Series.


```python
import pandas as pd

calories = {"day1": 420, "day2": 380, "day3": 390}

myvar = pd.Series(calories)

print(myvar)
```

```python
import pandas as pd

df = pd.read_csv('data.csv')
df
```

```python
# Overview of the DataFrame
print(df.info())

# Summary statistics
print(df.describe())

# First few rows
print(df.head())

# Shape (rows, columns)
print(df.shape)
```

```python
# Select one column
print(df['Duration'])

# Select multiple columns
print(df[['Duration', 'Calories']])

# Select specific row by index
print(df.loc[3])  # 4th row

# Select row and column
print(df.loc[3, 'Calories'])
```

```python
df.head(15) 
```

```python
df.tail(5)
```

```python
# Workouts longer than 45 minutes
print(df[df['Duration'] > 45])

# Workouts where Maxpulse is over 140
print(df[df['Maxpulse'] > 140])

# Workouts where Duration is 60 and Calories > 350
print(df[(df['Duration'] == 60) & (df['Calories'] > 350)])
```

```python
# Sort by Calories (ascending)
print(df.sort_values(by='Calories'))

# Sort by Maxpulse (descending)
print(df.sort_values(by='Maxpulse', ascending=False))
```

```python
# Add a new column: Calories burned per minute
df['CalPerMin'] = df['Calories'] / df['Duration']

# Modify existing column
df['Duration'] = df['Duration'] + 5  # Adding 5 to every duration
```

```python
# Drop the new column (without modifying the original df)
df.drop(columns=['CalPerMin'], inplace=True)

# Drop a specific row
df.drop(index=2, inplace=True)  # Remove 3rd row
```

```python
df
```

```python
# Average Calories burned grouped by Duration
print(df.groupby('Duration')['Calories'].mean())

# Multiple aggregations
print(df.groupby('Duration').agg({
    'Pulse': 'mean',
    'Maxpulse': 'max',
    'Calories': 'sum'
}))
```

```python
print(df.isnull().sum())  # Check for NaNs

# # Drop missing values
df.dropna(inplace=True)

# # Fill missing values
df.fillna(0, inplace=True)
df
```

```python
# Save to CSV
df.to_csv('workout_data.csv', index=False)

# Read from CSV
df_loaded = pd.read_csv('workout_data.csv')
```

```python
import matplotlib.pyplot as plt

# Plot Duration vs Calories
df.plot(kind='scatter', x='Duration', y='Calories')
plt.show()
```

```python
df.corr()
```

1. The corr() method ignores "not numeric" columns.

2. The Result of the corr() method is a table with a lot of numbers that represents how well the relationship is between two columns.
3. The number varies from -1 to 1.
4. 1 means that there is a 1 to 1 relationship (a perfect correlation), and for this data set, each time a value went up in the first column, the other one went up as well.
5.  0.9 is also a good relationship, and if you increase one value, the other will probably increase as well.

6. -0.9 would be just as good relationship as 0.9, but if you increase one value, the other will probably go down.

7. 0.2 means NOT a good relationship, meaning that if one value goes up does not mean that the other will.

8. What is a good correlation? It depends on the use, but I think it is safe to say you have to have at least 0.6 (or -0.6) to call it a good correlation.
